```python
# Importing the csv file for young surviving mice (Created in R)
import pandas as pd

res = pd.read_csv("res_young_surv_df.csv", index_col=0)
res.head()
```

```python
# Filtering significant genes
sig = res[
    (res["padj"] < 0.05) &
    (abs(res["log2FoldChange"]) > 1)
]

sig_genes = sig.index.tolist()
sig_genes = [g.split(".")[0] for g in sig_genes]  # Ensembl-Versionen entfernen
sig_genes[:10]
```

```python
# Applying KEGG analysis
!pip install gseapy mygene

import gseapy as gp

enr = gp.enrichr(
    gene_list=sig_genes,
    gene_sets="KEGG_2019_Mouse",
    cutoff=0.05
)

enr.results.head()
```

```python
# Filtering pathways for keywords 
kegg = enr.results  # aus gseapy
kegg[kegg['Term'].str.contains("FoxO|proteasome|ubiquitin|cardiac|circadian rhythm|notch signaling|inositol phosphate metabolism", case=False)]
```
